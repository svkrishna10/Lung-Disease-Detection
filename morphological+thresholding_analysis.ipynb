### 1. Global Thresholding using Mean value

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt

def apply_global_threshold(image):
    mean_value = np.mean(image)
    _, thresholded_image = cv2.threshold(image, mean_value, 255, cv2.THRESH_BINARY)
    return thresholded_image

# Folder path containing images
folder_path = 'D:/Amrita/3rdYr/Digital Image Processing/covid'

# Create a subplot with a single row and columns equal to the number of images
plt.figure(figsize=(15, 5))  # Adjust the figure size as needed

# Loop through all images in the folder
for i, filename in enumerate(os.listdir(folder_path)):
    if filename.endswith(('.jpg', '.jpeg', '.png')):  # Assuming image files have these extensions
        # Load the image
        image_path = os.path.join(folder_path, filename)
        original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Apply global thresholding using mean value
        thresholded_image = apply_global_threshold(original_image)

        # Display the original and thresholded images in one row
        plt.subplot(1, 6, 2 * i + 1)
        plt.imshow(original_image, cmap='gray')
        plt.title(f'Original - {filename}')
        plt.axis('off')

        plt.subplot(1, 6, 2 * i + 2)
        plt.imshow(thresholded_image, cmap='gray')
        plt.title(f'Thresholded - {filename}')
        plt.axis('off')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
# Function for region growing
def region_growing(image, seed):
    
    h, w = image.shape
    segmented = np.zeros_like(image, dtype=np.uint8)
    
    threshold = 20
    
    # Queue for pixel coordinates
    queue = [seed]
    
    # Seed pixel intensity
    seed_intensity = image[seed]
    
    while queue:
        current_point = queue.pop(0)
        
        # Check if the current point is within the image bounds
        if (0 <= current_point[0] < h) and (0 <= current_point[1] < w):
            # Check if the pixel is not already segmented and its intensity is similar to the seed
            if segmented[current_point] == 0 and abs(int(image[current_point]) - int(seed_intensity)) < threshold:
                segmented[current_point] = 255
                
                # Add neighbors to the queue
                queue.append((current_point[0] + 1, current_point[1]))
                queue.append((current_point[0] - 1, current_point[1]))
                queue.append((current_point[0], current_point[1] + 1))
                queue.append((current_point[0], current_point[1] - 1))
                queue.append((current_point[0] + 1, current_point[1]+1))
                queue.append((current_point[0] - 1, current_point[1]-1))
                queue.append((current_point[0]+1, current_point[1] - 1))
                queue.append((current_point[0]-1, current_point[1] + 1))
    
    return segmented

# Seed point (you can manually select a seed point or use some algorithm to find it)
seed_point = (150,250)


folder_path = 'D:/Amrita/3rdYr/Digital Image Processing/v_pneumonia'

# Create a subplot with a single row and columns equal to the number of images
plt.figure(figsize=(15, 5))  # Adjust the figure size as needed

# Loop through all images in the folder
for i, filename in enumerate(os.listdir(folder_path)):
    if filename.endswith(('.jpg', '.jpeg', '.png')):  # Assuming image files have these extensions
        # Load the image
        image_path = os.path.join(folder_path, filename)
        original_image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

        # Apply region growing
        segmented_image = region_growing(original_image, seed_point)

        # Display the original and adaptive thresholded images in one row
        plt.subplot(1, 6, 2 * i + 1)
        plt.imshow(original_image, cmap='gray')
        plt.title(f'Original - {filename}')
        plt.axis('off')

        plt.subplot(1, 6, 2 * i + 2)
        plt.imshow(segmented_image, cmap='gray')
        plt.title(f'Region Growing - {filename}')
        plt.axis('off')

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```
